# Cars DataBase

```python
cursor.execute("create database CarDataSet")
```

```python
cursor.execute("CREATE TABLE CARDATASET.CARSDATA(BUYING VARCHAR(10),MAINT VARCHAR(10), DOORS VARCHAR(19), PERSONS VARCHAR(10), LUG_BOOT VARCHAR(10), SAFETY VARCHAR(10), CLASS VARCHAR(10))")
```

Inserting data in bulk in to the table

```python
with open("car.data", "r") as carsdata:
    for line in carsdata:
        cleaned_data = [line.strip() for line in line.split(",")]
        cursor.execute(F"INSERT INTO CARDATASET.CARSDATA VALUES('{cleaned_data[0]}','{cleaned_data[1]}','{cleaned_data[2]}','{cleaned_data[3]}','{cleaned_data[4]}','{cleaned_data[5]}','{cleaned_data[6]}')")
mydb.commit()
```
